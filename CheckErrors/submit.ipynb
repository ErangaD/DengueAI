```python
feature_columns = ['city', 'reanalysis_specific_humidity_g_per_kg',
                'reanalysis_dew_point_temp_k',
                'station_avg_temp_c', 'precipitation_amt_mm', 'ndvi_ne', 'ndvi_nw', 'ndvi_se','ndvi_sw', 'weekofyear']

#ndvi = density of green
#precipitation = rain amount

df = pd.read_csv('dengue_features_train.csv', usecols=feature_columns)
```

```python
df['station_avg_temp_c'] = df['station_avg_temp_c'].rolling(window=3, center= True, min_periods=(1)).mean()
df['precipitation_amt_mm'] = df['precipitation_amt_mm'].rolling(window=3, center= True, min_periods=(1)).median()
# df = df.fillna(df.mean())
df['ndvi_ne']=df['ndvi_ne'].fillna(df['ndvi_nw'])
df['ndvi_nw']=df['ndvi_nw'].fillna(df['ndvi_se'])
df['ndvi_se']=df['ndvi_se'].fillna(df['ndvi_sw'])
df['ndvi_sw']=df['ndvi_sw'].fillna(df['ndvi_ne'])
df = df.fillna(df.mean())
```

```python
# df['week_start_date'] = pd.to_datetime(df['week_start_date'])
# df['week_start_date'] = df['week_start_date'].dt.week
```

```python
m = {'sj': 0, 'iq': 1}
df['city'] = df['city'].map(m)
```

```python
label_columns = ['total_cases']
labels = pd.read_csv('dengue_labels_train.csv', usecols=label_columns)
```
