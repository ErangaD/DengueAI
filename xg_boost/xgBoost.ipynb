```python
import pandas as pd
```

```python
X_sj = pd.read_csv('../X_sj.csv', index_col=0)
X_iq = pd.read_csv('../X_iq.csv', index_col=0)
X_sj
```

```python
y_sj = pd.read_csv('../y_sj.csv', header=None, index_col=0)
y_iq = pd.read_csv('../y_iq.csv', header=None, index_col=0)
y_sj
```

```python
X_test_sj = pd.read_csv('../X_test_sj.csv', index_col=0)
X_test_iq = pd.read_csv('../X_test_iq.csv', index_col=0)
```

```python
import xgboost as xgb

xgdmat_sj = xgb.DMatrix(X_sj, y_sj)
xgdmat_iq = xgb.DMatrix(X_iq, y_iq)
```

```python
params = {'eta': 0.01, 'seed':1, 'subsample': 0.8, 'colsample_bytree': 0.8, 
            'max_depth':6, 'min_child_weight':2, 'n_estimators': 100} 
# Grid Search CV optimized settings
```

```python
final_gb_sj = xgb.train(params, xgdmat_sj, num_boost_round = 2163)
# final_gb_iq = xgb.train(params, xgdmat_iq, num_boost_round = 2163)
```

```python
final_gb_iq = xgb.train(params, xgdmat_iq, num_boost_round = 2163)
X_test_iq
```

```python
testdmat_sj = xgb.DMatrix(X_test_sj)
testdmat_iq = xgb.DMatrix(X_test_iq)
```

```python
y_pred_sj = final_gb_sj.predict(testdmat_sj)
y_pred_iq = final_gb_iq.predict(testdmat_iq)
```

```python
for pred in y_pred_sj:
    print(int(round(pred)))
```

```python
for pred in y_pred_iq:
    print(int(round(pred)))
```
