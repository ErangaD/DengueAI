```python
import  pandas as  pd  # data processing, CSV file I/O (e.g. pd.read_csv)# data  
import warnings
warnings.filterwarnings("ignore")
```

```python
train_features = pd.read_csv('./../original data_set/dengue_features_train.csv')
train_labels = pd.read_csv('./../original data_set/dengue_labels_train.csv')
test_features = pd.read_csv('./../original data_set/dengue_features_test.csv')
train_features
```

```python
train_features.fillna(method='bfill', inplace=True)

test_features.fillna(method='bfill', inplace=True)

train_labels_sj = train_labels[train_labels['city'] == 'sj'].total_cases
train_labels_iq = train_labels[train_labels['city'] == 'iq'].total_cases

# dropping date column (inforation contained in year + week of year)
train_features = train_features.drop(train_features.columns[[3]], axis=1)

test_features = test_features.drop(test_features.columns[[3]], axis=1)
```

```python
df_sj = train_features[train_features['city'] == 'sj']
df_iq = train_features[train_features['city'] == 'iq']
dt_sj= test_features[test_features['city'] == 'sj']
dt_iq =test_features[test_features['city'] == 'iq']


test_features
```

```python
df_sj = df_sj.drop(df_sj.columns[[0]], axis=1)
df_iq =  df_iq.drop(df_iq.columns[[0]], axis=1)
dt_sj=  dt_sj.drop(dt_sj.columns[[0]], axis=1)
dt_iq = dt_iq.drop(dt_iq.columns[[0]], axis=1)
dt_iq
```

```python
threshold = 0.2

def cleanData(df):
    list_col=[]
    for col,val in df.items():
        if val<0.2:
            list_col.append(col)
    return list_col
sj_rem_col=cleanData(df_sj.std())
iq_rem_col=cleanData(df_iq.std())

df_sj = df_sj.drop(sj_rem_col, axis=1)
dt_sj = dt_sj.drop(sj_rem_col, axis=1)

df_iq = df_iq.drop(iq_rem_col, axis=1)
dt_iq = dt_iq.drop(iq_rem_col, axis=1)
```
